```python
import pandas as pd
```

```python
df_1 = pd.read_csv("2015.csv")
df_2 =  pd.read_csv("2016.csv")
```

```python
global_happiness = pd.merge(df_1,df_2, on= "Country")
global_happiness = global_happiness.rename(columns={"Happiness Rank_x" :"Happiness Rank_2015"})
```

```python
nuevos_nombres = []
for i in global_happiness.columns:
    if "_x" in i:
        nuevos_nombres.append(i.replace("_x","_2015"))
    elif "_y" in i:
        nuevos_nombres.append(i.replace("_y","_2016"))
    else:
        nuevos_nombres.append(i)
```

```python
global_happiness.columns = nuevos_nombres
global_happiness
```
