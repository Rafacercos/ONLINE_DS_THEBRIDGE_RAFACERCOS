3. Fija el eje X entre 0 y 5 y el eje Y entre -15 y 15

```python
fig , ax = plt.subplots(1,2)
x = np.linspace(0,5)
m_1 = 3
b_1 = 5
y = m_1*x + b_1
x = np.linspace(0,5)
ax[0].plot(x,y)
ax[0].scatter(0,b_1)
ax[0].set_ylim(-15,15)
m_2 = -3
b_2 = -5
y = m_2*x + b_2
x = np.linspace(0,5)
ax[1].plot(x,y)
ax[1].scatter(0,b_2)
ax[1].set_ylim(-15,15)  
```

4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
fig , ax = plt.subplots(1,2)
x = np.linspace(0,5)
m_1 = 3
b_1 = 5
y = m_1*x + b_1
x = np.linspace(0,5)
ax[0].set_xlabel("Eje x")
ax[0].set_ylabel("Eje y")
ax[0].plot(x,y)
ax[0].scatter(0,b_1)
ax[0].set_ylim(-15,15)
ax[0].set_title("Recta sube")
m_2 = -3
b_2 = -5
y = m_2*x + b_2
x = np.linspace(0,5)
ax[1].plot(x,y)
ax[1].scatter(0,b_2)
ax[1].set_ylim(-15,15) 
ax[1].set_xlabel("Eje x")
ax[1].set_ylabel("Eje y") 
ax[1].set_title("Recta baja")
plt.tight_layout() #Ajusta el espaciado entre gráficas automáticamente 
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0,5)
y = 3*x +5
plt.figure()
plt.subplot(1 ,2,1)
plt.plot(x,y)
plt.ylim(-15,15)
plt.xlabel("Eje x")
plt.ylabel("Eje y") 
plt.title("Recta sube")
y = -3*x -5
plt.subplot(1 ,2,2)
plt.plot(x,y)
plt.ylim(-15,15)
plt.xlabel("Eje x")
plt.ylabel("Eje y") 
plt.title("Recta sube")
plt.tight_layout() 
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

Pista: https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html

```python
lista_alturas= np.random.randint(150,190,300)
dict_2 = {"Alturas":lista_alturas }
ejer_19 = pd.DataFrame(dict_2)
plt.figure(figsize=(15,8))
plt.hist(ejer_19, bins = len(ejer_19["Alturas"]))
plt.xticks(ticks= ejer_19["Alturas"].unique())
plt.ylabel("Numero de personas")
plt.xlabel("Alturas")
```
